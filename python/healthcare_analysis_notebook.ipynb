# Healthcare Encounter Analytics

Exploration, modeling, and interpretation for the cleaned healthcare dataset.

```
import pandas as pd
from pathlib import Path

df = pd.read_csv('healthcare_dataset_cleaned.csv')
df.head()
```

```
df.shape, df.isna().sum().sort_values(ascending=False).head(10)
```

```
df['medical_condition'].value_counts()
```

```
df.groupby('medical_condition').agg(encounters=('encounter_id','count'), avg_billing=('billing_amount','mean'), avg_los_days=('length_of_stay_days','mean')).sort_values('avg_billing', ascending=False)
```

```
from sklearn.model_selection import train_test_split
from sklearn.dummy import DummyRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score, mean_absolute_error

features = ['age','gender','blood_type','medical_condition','insurance_provider','admission_type','medication','test_results','length_of_stay_days','admission_year']
reg = df[df.billing_amount.notna()].sample(n=min(20000, df.billing_amount.notna().sum()), random_state=42)
X = pd.get_dummies(reg[features])
y = reg['billing_amount']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.25, random_state=42)
model = DecisionTreeRegressor(max_depth=5, min_samples_leaf=200, random_state=42)
model.fit(X_train, y_train)
pred = model.predict(X_test)
{'r2': r2_score(y_test, pred), 'mae': mean_absolute_error(y_test, pred)}
```

## Interpretation

The available data produce a weak predictive signal. This is a valid analytic finding to be stated directly in the README.
